In this section, some intermediate techniques to get better prediction are explored.

```python
import kagglehub
import pandas as pd
import os #is used to import Operating system, is a built-in Python library that allows your code to interact with the operating system your script is running on (whether that's Linux in Google Colab, Windows on your laptop, or macOS)
path = kagglehub.dataset_download("dansbecker/melbourne-housing-snapshot")#is a folder
print("Files in folder:", os.listdir(path))#list all files in the folder
file_path = os.path.join(path, 'melb_data.csv')#simple operation to get the correct path to the file
data = pd.read_csv(file_path)#load the data as a dataframe
print(data.head())#prints the first five rows and the title of each column
data.describe()# gives statical description of the data
```

```python
from sklearn.model_selection import train_test_split
y = data.Price
melb_predictors = data.drop(['Price'], axis = 1)
X = melb_predictors.select_dtypes(exclude = ['object'])
X_train, X_valid, y_train, y_valid = train_test_split(X, y,train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_dataset(X_train, X_valid, y_train, y_valid):
  model = RandomForestRegressor(n_estimators= 10, random_state = 0)
  model.fit(X_train, y_train)
  predics = model.predict(X_valid)
  return mean_absolute_error(y_valid, predics)




```

In this section the performance of the model with the approach of dropping the columns with missing values is been evaluated by calculating its Mean Absolute Error.


```python
columns_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()]
reduced_X_train = X_train.drop(columns_with_missing, axis = 1)
reduced_X_valid = X_valid.drop(columns_with_missing, axis = 1)
print("MAE from Approach 1 (Drop columns with missing values):")
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

In this section the performance of the model with the approach of imputation is been evaluated by calculating its Mean Absolute Error.

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
print("MAE from Approach 2 (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

In this section the performance of the model with the approach of extension to imputation is been evaluated by calculating its Mean Absolute Error.

```python
X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()

for col in columns_with_missing:
  X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()
  X_valid_plus[col + '_was_missing'] = X_valid_plus[col].isnull()
my_imputer = SimpleImputer()
imputed_X_train_plus = pd.DataFrame(my_imputer.fit_transform(X_train_plus))
imputed_X_valid_plus = pd.DataFrame(my_imputer.transform(X_valid_plus))
imputed_X_train_plus.columns = X_train_plus.columns
imputed_X_valid_plus.columns = X_valid_plus.columns
print("MAE from Approach 2 (an Extension to Imputation):")
print(score_dataset(imputed_X_train_plus,imputed_X_valid_plus, y_train, y_valid))
```

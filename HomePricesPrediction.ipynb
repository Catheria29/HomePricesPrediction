In this part, we are going to explore how to use data, to get a better model that works accurately (at least mostly) not only for the "in sample"(training data and validation data are the same) but also for the "out_of_sample data" (a portionof data was isolated from the training data using: train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0), this part is called validation data. it will be used to test the model and calculate the mean absolute error

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=0)#returns numpu arrays
melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X,train_y)
val_prediction = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y,val_prediction))
print(val_y.describe())


```

From this result , MAE = 246418.66863033874 and the average of the house price in the validation data is 1.076293e+06, so the error is about quarter of the average home value.

The phenomenon called **overfitting** (The "Memorizer"), where a model matches the training data almost perfectly, but does poorly in validation and other new data. ***Example***: When a tree has too many leaves, it basically "memorizes" the specific houses it was trained on.
When a model fails to capture important distinctions and patterns in the data, so it performs poorly even in training data, that is called **underfitting** (The "Generalizer"). ***Example***: When a tree is too shallow (too few leaves), it groups houses together that don't actually belong together.

**Investigate and optimize the decicion tree** **model** **bold text**

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
  model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state = 0)
  model.fit(train_X,train_y)
  predicts_val = model.predict(val_X)
  mae = mean_absolute_error(val_y, predicts_val)
  return mae
```

```python
for i in [490,500,550, 600, 5000]:
  my_mae = get_mae(i, train_X, val_X, train_y, val_y)
  print("Max leaf nodes = %d \t\t Mean Absolute Error = %d"%(i,my_mae))
```

We can ddeduce that the minimum Max leaf nodes = 500 		 Mean Absolute Error = 231301, thus, this can be used to fix our model accuracy if decision tree is still the choosen model.

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=500, random_state = 0)
final_model.fit(X,y)
```

We'll look at the random forest as an example of better model to improve the accuracy of our prediction.

The random forest uses many trees, and it makes a prediction by averaging the predictions of each component tree. It generally has much better predictive accuracy than a single decision tree and it works well with default parameters.



```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y,melb_preds))
```
